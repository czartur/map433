#**Q.1**
La densité $p_\theta$ par rapport à la mesure $Leb^{⊗ n}$ est donnée par:

$$p_\theta(x_1,\,x_2,\,\cdots ,\,x_n) = \prod^{n}_{i=1}\mathcal{N}(\beta_1+\beta_2 t_i;\, \sigma^2)[x_i] = \frac{1}{(2\pi \sigma^2)^{\frac{n}{2}}}\exp \left(-0,5\sigma^{-2}\sum^{n}_{i=1}(x_i - \beta_1 - \beta_2 t_i)^2\right)$$

Alors, de l'expression de $p_\theta$, on déduit que la vraisemblance est donnée par:

$$\theta \to L_n(\theta,\, X_1,\,\cdots ,\, X_n) = \frac{1}{(2\pi \sigma^2)^{\frac{n}{2}}}\exp \left(-0,5\sigma^{-2}\sum^{n}_{i=1}(x_i - \beta_1 - \beta_2 t_i)^2\right)$$

Si on note les estimateus des vraisemblance de $\beta_1$ et $\beta_2$ par $\beta^*_1$ et $\beta^*_2$, on voit que ces estimateurs sont obtenus en minimisant l'argument de l'exponentiel, c'est à dire, en minimisant l'expression:

$$\sum^{n}_{i=1}(x_i - \beta_1 - \beta_2 t_i)^2$$

Mais cet expression est égal à l'expresion de $S(\beta_1,\, \beta_2)$, alors les estimateurs de maximum de vraisemblance coincident avec les estimateurs des moindres carrés, c'est à dire: 

$$\hat{\beta_1} = \beta_1^*\,\,\,\, \textrm{ et }\,\,\,\, \hat{\beta_2} = \beta_2^*$$

Les expressions pour calculer $\beta_1^*$ et $\beta_2^*$ ont été vus en PC2 et, si on note:

$$\textbf{T} := [\textbf{1}\;\;\;\textbf{t}]\in\mathbb{R}^{n\times 2}\;\;\;\textrm{ et }\;\;\;\hat{\beta} := [\hat{\beta_1}\;\;\;\hat{\beta_2}]^{T}\in\mathbb{R}^2$$

Comme il existe au moins deux indices $i\neq j$ tels que $t_i\neq t_j$, on a que $T^TT$ est invertible et $\hat{\beta}$ est exprimé par:

$$\hat{\beta} = (T^TT)^{-1}T^TX$$

Si on développe l'expression si dessus, on obtien:

$$\hat{\beta_2} = \frac{\sum^{n}_{i=1}(t_i-\bar{t})X_i}{\sum^{n}_{i=1}(t_i-\bar{t})^2} \;\;\;\textrm{ et }\;\;\; \hat{\beta_1} = \bar{X} - \hat{\beta_2}\bar{t}$$

#**Q.3**
On estime les paramètres $\hat{β_1}$, $\hat{β_2}$ et $\hat{σ}^{2}$ en utilisant les données d’anomalies sur des intervalles de temps de 30 ans (1880-1910), (1890-1920), etc... à partir des formules trouvées à la question 1.

> Les estimations des paramètres sont representés à partir d'un DataFrame pour chaque intervalle de 30 ans.

```python
 def estimate(year):
  
    #30 years range temperature mean vector
    X = df['J-D'][year-1880:year-1880+n]
    Xm = np.array([np.mean(X)]*n)

    #years mean vector
    T = np.array([year+i for i in range(n)])
    Tm = np.array([np.mean(T)]*n)
    
    #formulas for b1, b2
    b2 = X.dot(T-Tm) /(T-Tm).dot(T-Tm)
    b1 = Xm[0] - b2*Tm[0]
    
    #formulas for sigma^2 
    X_ = b1*np.ones(n) + b2*T
    E = X-X_
    sigmaCarre = E.dot(E)/(n-2) 
      
    return sigmaCarre, b1, b2

estimations = pd.DataFrame([(str(year)+"~"+str(year+n),)+estimate(year) for year in range(1880, 2000, 10)], columns = ['Year-Range', 'sigma^2', 'B1', 'B2'])
estimations
```

#**Q.4**

Conformement à la question 1 de la PC2, on retrouve les formules suivantes pour les intervalles de confiance de niveau $1-α$ :

* $σ^2$ (quantiles de la loi chi-deux) : $\left[ n \frac{\hat{σ^2}}{q_{1-α/2}^{n-2}}, n \frac{\hat{σ^2}}{q_{α/2}^{n-2}}  \right]$

* $β_1$ (quantiles de la loi t-student) : $\left[ ᵦ_1 - σ_1 t_{1-α/2}^{n-2}, ᵦ_1 - σ_1 t_{α/2}^{n-2}\right] $

* $β_2$ (quantiles de la loi t-student) : $\left[ ᵦ_2 - σ_2 t_{1-α/2}^{n-2}, ᵦ_2 - σ_2 t_{α/2}^{n-2}\right] $

> Les intervalles de confiance (niveau 0.95) des paramètres sont representés à partir d'un DataFrame pour chaque intervalle de 30 ans.


```python
def intervals(year, alpha=0.05):
    #get parameters estimation
    sigmaCarre, b1, b2 = estimate(year)
    X = df['J-D'][year-1880:year-1880+n]
    T = np.array([year+i for i in range(n)])
    Tm = np.array([np.mean(T)]*n)
    
    #quantiles
    ql = t.ppf(alpha/2, n-2)
    qr = t.ppf(1-alpha/2, n-2)
    ql_chi2 = chi2.ppf(alpha/2, n-2)
    qr_chi2 = chi2.ppf(1-alpha/2, n-2)

    #sigma1&2 formulas
    sigma1 = np.sqrt(sigmaCarre*T.dot(T)/(n*(T-Tm).dot(T-Tm)))
    sigma2 = np.sqrt(sigmaCarre/(T-Tm).dot(T-Tm))

    
    #formules pour les intervalles de confiance
    [a0, b0] = [(n-2)*sigmaCarre/qr_chi2, (n-2)*sigmaCarre/ql_chi2]
    [a1, b1] = [b1 - sigma1*qr, b1 - sigma1*ql] #B1 in [a1,b1]
    [a2, b2] = [b2 - sigma2*qr, b2 - sigma2*ql] #B2 in [a2,b2]
    
    return "["+str(a0)+",    "+str(b0)+"]", "["+str(a1)+",    "+str(b1)+"]", "["+str(a2)+",    "+str(b2)+"]"

intervalles = pd.DataFrame([(str(year)+"~"+str(year+n),) + intervals(year) for year in range(1880, 2000, 10)], columns = ['Year-Range', 'sigma^2', 'B1', 'B2'])
intervalles
```

$$ \hat{\sigma} \, t_{1-α/2}^{n-2} \sqrt{1+ 1/n + \frac{(ti-\bar{t})^2}{\sum_{i=1}^{n}(ti-\bar{t})^2}} $$

Puis on visualise en superposition du graphe des observations les différentes droites de régression $\hat{X_i} = \hat{β_1}+\hat{β_2}t_i$ et les intervalles de confiance de prédiction de niveau de couverture 0.95 ($α = 0.05$).

> Choisissez entre plotter le graphique entière ou seulement un intervalle de 30 ans en mettant une année quelconque entre 1800 et 1990 comme paramètre de la fonction.

```python
def plot_prediction(year=0, alpha = 0.05):

  if year == 0:
    for year in range(1880, 2000, 10):
      plot_prediction(year, alpha=alpha)
    plt.plot(df.Year, df['J-D'])

  else :
    #getting data
    X = df['J-D'][year-1880:year-1880+n]
    T = np.array([year+i for i in range(n)])

    #linear regression
    sigmaCarre, b1, b2 = estimate(year)
    X_ = b1*np.ones(n) + b2*T

    #Confidence interval of prediction
    quantile = t.ppf(1-alpha/2, n-2)
    sigma = np.sqrt(sigmaCarre)
    IC = quantile*sigma*np.array([np.sqrt(1+1/n+((year + i - T.mean())**2)/(T-np.mean(T)).dot(T-np.mean(T))) for i in range(n)])
    #plot 30-years range observations
    plt.plot(df.Year[year-1880:year-1880+n], df['J-D'][year-1880:year-1880+n])
    
    # plot 30-years range prediction
    plt.plot(T, X_)

    # plot confidence interval curves for this range
    plt.fill_between(T, (X_ - IC), (X_+ IC), color='b', alpha=0.05)
    

  

#plot_prediction()
plot_prediction(1970)
```

#**Q.6**
Test bilatère $H_0$ : $β_2 = 0$, contre $H_1$ : $β_2 \neq 0$

On définit notre région de rejet de la façon suivante :

$ R_{\alpha} = \{ Z | \, \frac{\hat{\beta_2}}{\hat{\sigma_2}} > t_{1-\alpha}^{n-2}\} $

On voit bien que pour $ \alpha < \alpha' $:
* $ R_{\alpha} \subset R_{\alpha'}$
* Le teste définit pour chaque région $R_{\alpha}$ est de niveau α :

La p-valeur est donc la solution de l'équation suivante : 

$$ \frac{\hat{\beta_2}}{\hat{\sigma_2}} = t_{1-\alpha}^{n-2} $$

Soit encore :   

$$ p-value = 1 - \mathbb{P}(Y \leq \frac{\hat{\beta_2}}{\hat{\sigma_2}})$$

où $Y$ suite la loi T-student de $n-2$ degrées.


On montre les p-valeurs pour chaque intervalle de 30 ans et on plotte en superposition du graphique des observations des points rouges qui répresentent exactement chaque valeur de p-valeur. 

```python
# Rejection region --->  {Z | T(Z)=b2/sigma2 > t(1-a)}  => p-value = 1 - P( student(n-2) <= b2/sigma2 )

def p_value(year):
  #getting data
  T = np.array([year+i for i in range(n)])
  Tm = np.mean(T)

  #get estimation for the observation Z
  sigmaCarre, b1, b2 = estimate(year)
  sigma2 = np.sqrt(sigmaCarre/(T-Tm).dot(T-Tm))
  
  return 1-t.cdf(b2/sigma2, n-2)

for year in range(1880, 2000, 10):
  plt.scatter(year+15, p_value(year), color="r")

plot_prediction()
plt.plot(df.Year, df['J-D'], color="b", alpha=0.5)
plt.plot()

p_val = pd.DataFrame([(str(year)+"~"+str(year+n),p_value(year)) for year in range(1880, 2000,10)], columns = ['Year-range', 'p-valeur'])
p_val
```
